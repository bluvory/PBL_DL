# 1. Linear Regression : 조금 더 어려운 데이터

### 1.1 데이터 준비

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
boston.feature_names
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
# 데이터 분할
from sklearn.model_selection import train_test_split
# 전체에서 train : test = 8 : 2
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=1)
```

```
train_x.shape
```

**먼저 Network을 그려봅시다!!!**

### 1.2 모델링

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
model = keras.models.Sequential()

# ③ 블록 준비
lc = keras.layers.Dense(1, input_shape = (13,))   # train_x.shape[1] = 13

# ④ model에 블럭 추가하기
model.add(lc)

# ⑤ Compile
model.compile(loss='mse', optimizer='adam')
```

```
# 모델을 열어 봅시다.
model.summary()
```

```
# 2. 학습시키기
model.fit(train_x,train_y, epochs=10)  # mse 줄어들면서 parameter 조정
```

```
# 3. 예측하기
test_pred = model.predict(test_x)
```

```
# 4. 평가하기
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test_y, test_pred)
```

# 2. Logistic Regression : 조금 어려운 데이터

### 2.1 데이터 준비

```
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

# 모양 확인해볼 것
x.shape, y.shape
```

```
# 데이터 분할
from sklearn.model_selection import train_test_split
# 전체에서 train : test = 8 : 2
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=1)
```

```
train_x.shape
```

**먼제 Network를 그려봅시다!!!**

### 2.2 모델링

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()


# ② 모델 정의
model = keras.models.Sequential()


# ③ 블록 준비 
lc = keras.layers.Dense(1, input_shape=(30,) activation = 'sigmoid')


# ④ model에 블럭 추가하기
model.add(lc)


# ⑤ 컴파일
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

```

```
# 모델을 열어 봅시다.
model.summary()

```

```
# 2. 학습시키기
model.fit(train_x, train_y, epochs = 10)
```

```
# 3. 예측하기
test_pred = model.predict(test_x)
```

```
# 4. 평가하기
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```
test_pred
```

```
test_pred2 = (test_pred >= 0.5).astype(int)
```

```
accuracy_score(test_y, test_pred2)
```

```
print(classification_report(test_y, test_pred2))
```
