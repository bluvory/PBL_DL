## 1. Quick Linear Regression!

```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```
x = np.array(range(0,20)) 
y = x * 2 -1 
print(x)
print(y)
```

```
plt.plot(x,y, 'bo')
plt.show()
```

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# ③ 블록 준비 : 인풋을 받아, weight을 곱하고, bias를 더해주는거 까지 한 번에 해주는 거
# 1, : 아웃풋이 1개
# input_shape=(1,) : 인풋이 하나
layer1 = keras.layers.Dense(1,input_shape=(1,))

# ④ model에 블럭 추가하기
model.add(layer1)

# ⑤ 컴파일
model.compile(loss = 'mse', optimizer = 'adam')  
```

```
# 2. 학습시키기
# epochs : 몇번 학습?
model.fit(x[:15], y[:15], epochs=10)
```

```
# 3. 예측하기
print(y[15:])
print(model.predict(x[15:]))
```

## 3. Quick Logistic Regression!

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```
x = np.array(range(0,20)) 
y = np.array([0]*10 + [1]*10)
print(x)
print(y)
```

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

# ③ 블록 준비 : 인풋을 받아, weight을 곱하고, bias를 더해주는거 까지 한 번에 해주는 거
# 1, : 아웃풋이 1개
# input_shape=(1,) : 인풋이 하나
layer1 = keras.layers.Dense(1, input_shape=(1,), activation='sigmoid')

# ④ model에 블럭 추가하기
model.add(layer1)

# ⑤ 컴파일
# 이진분류 loss function = binary_crossentropy
# 분류문제 metrics = accuracy
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```
# 2. 학습시키기
model.fit(x[:15], y[:15], epochs=10, verbose=1)
```

```
# 3. 예측하기
print(y[15:])
print(model.predict(x[15:]))
```
