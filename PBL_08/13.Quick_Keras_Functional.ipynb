## 1. Quick Linear Regression!

```
import tensorflow as tf
import tensorflow.keras as keras

import numpy as np
```

```
x = np.array(range(0,20)) 
y = x * 2 -1 
print(x)
print(y)
```

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# 레이어들을 사슬로 연결하 듯이 연결! 모든 레이어 다 선언하고, 모든 레이어 연결까지 써줘야 한다.
input_layer = keras.layers.Input(shape=(1,))  #Input layer
output_layer = keras.layers.Dense(1)(input_layer)  # input layer에 out으로 연결

# ③ 모델의 시작과 끝을 지정
model = keras.models.Model(inputs=input_layer, outputs=output_layer)

# ⑤ 컴파일
model.compile(loss = 'mse', optimizer = 'adam')
```

```
# 2. 학습시키기
model.fit(x, y, epochs=10, verbose=1)
```

```
# 3. 예측하기

print(y[15:])
print(model.predict(x[15:]))
```

```
# 패키지 로딩부터
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np

```

```
x = np.array(range(0,20)) 
y = x * (-3) + 10 
print(x)
print(y)
```

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# 레이어들을 사슬로 연결하 듯이 연결! 모든 레이어 다 선언하고, 모든 레이어 연결까지 써줘야 한다.
il = keras.layers.Input(shape=(1,))
ol = keras.layers.Dense(1)(il)

# ③ 모델의 시작과 끝을 지정
model = keras.models.Model(inputs = il, outputs = ol)

# ⑤ 컴파일
model.compile(loss = 'mse', optimizer = 'adam')
```

```
# 2. 학습시키기
model.fit(x,y,epochs=10, verbose=1)

```

```
# 3. 예측하기
print(y[15:])
print(model.predict(x[15:]))
```

## 3. Quick Logistic Regression!

```
import tensorflow as tf
import tensorflow.keras as keras

import numpy as np
```

```
x = np.array(range(0,20)) 
y = np.array([0]*10 + [1]*10)
print(x)
print(y)
```

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# 레이어들을 사슬로 연결하 듯이 연결! 모든 레이어 다 선언하고, 모든 레이어 연결까지 써줘야 한다.
input_layer = keras.layers.Input(shape=(1,))
output_layer = keras.layers.Dense(1, activation='sigmoid')(input_layer)

# ③ 모델의 시작과 끝을 지정
model = keras.models.Model(inputs=input_layer, outputs=output_layer)

# ⑤ 컴파일
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```
# 2. 학습시키기
model.fit(x[:15], y[:15], epochs=10, verbose=1)
```

```
# 3. 예측하기
print(y[15:])
print(model.predict(x[15:]))
```

## 4. 실습 : 무작정 따라하기

```
# 패키지 로딩부터
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np

```

```
x = np.array(range(0,40)) 
y = np.array([0]*20 + [1]*20)
print(x)
print(y)
```

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# 레이어들을 사슬로 연결하 듯이 연결! 모든 레이어 다 선언하고, 모든 레이어 연결까지 써줘야 한다.
il = keras.layers.Input(shape=(1,))
ol = keras.layers.Dense(1, activation='sigmoid')(il)

# ③ 모델의 시작과 끝을 지정
model = keras.models.Model(inputs = il, outputs=ol)

# ⑤ 컴파일
model.compile(loss='binary_crossentropy', optimizer = 'adam')
```
