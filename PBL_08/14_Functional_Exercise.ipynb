```
import tensorflow as tf
import tensorflow.keras as keras

import numpy as np
import pandas as pd
```

# 1. Linear Regression : 조금 어려운 데이터

### 1.1 데이터 준비


```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
x = boston.data
y = boston.target

x.shape, y.shape
```

```
from sklearn.model_selection import train_test_split
# 먼저 전체에서 train : test = 8 : 2
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=1)
```

**모델링 전에 Network Diagram을 먼저 그려봅시다!!!**

### 1.2 모델링

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
# 레이어들을 사슬로 연결하 듯이 연결! 모든 레이어 다 선언하고, 모든 레이어 연결까지 써줘야 한다.
il = keras.layers.Input(shape=(13,))
ol = keras.layers.Dense(1)(il)

# ③ 모델의 시작과 끝을 지정
model = keras.models.Model(inputs = il, outputs = ol)

# ④ 컴파일
model.compile(loss = 'mse', optimizer = 'adam')   # mse = keras.loses.mse, adam = keras.optimizers.Adam(러닝메이트조정가능)
```

```
# 2. 학습시키기
model.fit(train_x, train_y, epochs = 10, batch_size = 64)   # batch size : 한번 학습할때 몇등분해서 학습하는가
```

```
# 3. 예측하기
test_pred = model.predict(test_x)
```

# 2. Logistic Regression : 조금 어려운 데이터

### 2.1 데이터 준비

```
from sklearn.datasets import load_breast_cancer
```

```
cancer = load_breast_cancer()
```

```
x = cancer.data
y = cancer.target

x.shape, y.shape
```

```
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=1)
```

**모델링 전에 Network Diagram을 먼저 그려봅시다!!!**

### 2.2 모델링

```
# 1. 모델 선언하기
# ①세션 비우기 : 기계적으로 수행하자.
keras.backend.clear_session()

# ② 모델 정의
il = keras.layers.Input(shape=(30,))
ol = keras.layers.Dense(1, activation = 'sigmoid')(il)    # sigmoid = keras.activations.sigmoid

# ③ 모델의 시작과 끝을 지정
model = keras.models.Model(inputs = il, outputs = ol)

# ④ 컴파일
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
# binary_crossentropy = keras.losses.binary_crossentropy, adam = keras.optimizers.Adam(), accrucy = keras.metrics.Accuracy()
```

```
# 2. 학습시키기
model.fit(train_x, train_y, epochs = 10, batch_size = 128)
```

```
# 3. 예측하기
test_pred = model.predict(test_x)
```
