```
# 모델 평가
import tensorflow as tf
import numpy as np

# 데이터 준비 
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# help(tf.keras.datasets.mnist.load_data)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

# one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# minmax normalize & reshape
x_train = x_train / 255.
x_test = x_test / 255.

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```
# 모델을 평가한다
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(28, 28)))

cell = tf.keras.layers.SimpleRNNCell(64)
model.add(tf.keras.layers.RNN(cell))

model.add(tf.keras.layers.Dense(10, activation="softmax"))
```

```
model.compile( optimizer="adam", loss="categorical_crossentropy", metrics=['accuracy'])
```

```
model.fit(x_train, y_train, batch_size=128, epochs=10,
          shuffle=True, validation_split=0.2)
```

```
import pandas as pd

url = "https://raw.githubusercontent.com/blackdew/DeepLearningZeroToAll/master/data-02-stock_daily.csv"
datasets = pd.read_csv(url, skiprows=1)
datasets.head()
```

```
print(type(datasets))
data = datasets.values
print(type(data))
print(data.shape)

# 데이터를 역순으로 최근 데이터를 맨 아래로 한다.
data = data[::-1]

# minmax normalize
# (X - min) / (max - min)
data = (data - data.min(axis=0)) / (data.max(axis=0) - data.min(axis=0))

print(type(data))
print(data.shape)
```

```
x_train, y_train = [], []
for i in range(len(data) - 7):
  x_train.append(data[i:i+7, :])
  y_train.append(data[i+7, -1])

x_train = np.array(x_train)
y_train = np.array(y_train)
print(x_train.shape, y_train.shape)
```

```
import tensorflow as tf

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
print(x_train.shape, x_test.shape)
print(x_test.shape, y_test.shape)

# one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# minmax normalize
x_train = x_train / 255.
x_test = x_test / 255.

# shape 확인
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```
#################################
# vanilla CNN
X1 = tf.keras.layers.Input(shape=(28, 28, 1))
# conv1
H = tf.keras.layers.Conv2D(filters=128, kernel_size=3, strides=1, padding='same')(X1)
H = tf.keras.layers.MaxPool2D(pool_size=2, strides=2, padding='same')(H)
H = tf.keras.layers.Activation('swish')(H)
# conv2
H = tf.keras.layers.Conv2D(128, 3, 1, 'same')(H)
H = tf.keras.layers.MaxPool2D(2, 2, 'same')(H)
H = tf.keras.layers.Activation('swish')(H)
# flatten
H = tf.keras.layers.Flatten()(H)
H = tf.keras.layers.Dense(100, activation="swish")(H)

#################################
# LSTM
X2 = tf.keras.layers.Input(shape=(28, 28))
H2 = tf.keras.layers.LSTM(64)(X2)
H2 = tf.keras.layers.Dense(50, activation='swish')(H2)

# concatenate
H3 = tf.keras.layers.Concatenate(axis=1)([H, H2])
H3 = tf.keras.layers.Dense(50, activation='swish')(H3)

Y = tf.keras.layers.Dense(10, activation="softmax")(H3)

model = tf.keras.models.Model([X1, X2], Y)
```

```
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=['accuracy']
)
```

```
model.fit([x_train.reshape(-1, 28, 28, 1), x_train], 
          y_train, 
          batch_size=256, epochs=10, shuffle=True,
          validation_split=0.2)
```
